## AI Supported Artworks (ChatGPT Responses)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Blank canvas
canvas = np.zeros((820, 820, 3), dtype=np.uint8)

# Draw colorful circles
for i in range(0, 800, 50):
    color = tuple(np.random.randint(0, 256, size=3).tolist())
    cv2.circle(canvas, (410, 410), i//2, color, thickness=3)

# Display
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
canvas = np.zeros((820, 820, 3), dtype=np.uint8)

# Draw intersecting random lines
for _ in range(50):
    pt1 = tuple(np.random.randint(0, 820, 2))
    pt2 = tuple(np.random.randint(0, 820, 2))
    color = tuple(np.random.randint(0, 256, size=3).tolist())
    cv2.line(canvas, pt1, pt2, color, thickness=np.random.randint(1,4))

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
canvas = np.zeros((820, 820, 3), dtype=np.uint8)

for y in range(0, 820, 40):
    for x in range(0, 820, 40):
        color = tuple(np.random.randint(0, 256, size=3).tolist())
        cv2.rectangle(canvas, (x,y), (x+40,y+40), color, -1)

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
canvas = np.zeros((820, 820, 3), dtype=np.uint8)

center = (410, 410)
for t in range(0, 360*5, 5):  # 5 turns
    angle = np.radians(t)
    r = 2 * t
    x = int(center[0] + r * np.cos(angle))
    y = int(center[1] + r * np.sin(angle))
    color = tuple(np.random.randint(0, 256, size=3).tolist())
    cv2.circle(canvas, (x,y), 5, color, -1)

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
